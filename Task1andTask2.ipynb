```python
import numpy as np
import torch
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# ----------------------------------
# FeatureExtractor Class
# ----------------------------------
class FeatureExtractor:
    def __init__(self, batch_size=32):
        """
        Initializes the FeatureExtractor with a pre-trained EfficientNet-B0 model.
        """
        # Load EfficientNet-B0 pre-trained on ImageNet
        weights = EfficientNet_B0_Weights.IMAGENET1K_V1
        self.model = efficientnet_b0(weights=weights)
        self.model = torch.nn.Sequential(*list(self.model.children())[:-1])  # Remove the FC layer
        self.model.eval()  # Set the model to evaluation mode

        # Define image transformation pipeline
        self.transform = weights.transforms()  # Use the predefined preprocessing pipeline
        self.batch_size = batch_size

    def extract(self, images):
        """
        Extracts feature representations from the EfficientNet-B0 model.
        Args:
            images (numpy.ndarray): Array of raw images (shape: N x H x W x C).
        Returns:
            numpy.ndarray: Array of feature vectors (shape: N x D).
        """
        # Validate input shape
        if len(images.shape) != 4 or images.shape[-1] not in [1, 3]:
            raise ValueError("Input should be a 4D numpy array with shape (N, H, W, C) where C is 1 or 3.")

        # Prepare dataset and DataLoader
        class ImageDataset(Dataset):
            def __init__(self, images, transform):
                self.images = images
                self.transform = transform

            def __len__(self):
                return len(self.images)

            def __getitem__(self, idx):
                img = self.images[idx]
                if img.shape[-1] == 1:  # If grayscale, repeat channels to make it RGB
                    img = np.repeat(img, 3, axis=-1)
                img = torch.tensor(img).permute(2, 0, 1)  # Convert to (C, H, W)
                return self.transform(img)

        dataset = ImageDataset(images, self.transform)
        dataloader = DataLoader(dataset, batch_size=self.batch_size, shuffle=False)

        # Extract features
        features = []
        with torch.no_grad():  # No gradients needed
            for batch in dataloader:
                batch = batch.float()  # Ensure the batch is in float format
                batch_features = self.model(batch).squeeze()  # Extract features
                features.append(batch_features.numpy())  # Convert to NumPy

        return np.vstack(features)  # Combine features into a single array



# ----------------------------------
# PrototypeClassifier Class
# ----------------------------------
class PrototypeClassifier:
    def __init__(self):
        """
        Initialize the classifier for Learning with Prototypes.
        """
        self.prototypes = {}  # Dictionary to hold prototypes for each class
        self.class_counts = {}  # Keep track of the number of samples for each class

    def euclidean_distance(self, x, y):
        """
        Calculate Euclidean distance between two vectors.
        Args:
            x (numpy.ndarray): First vector.
            y (numpy.ndarray): Second vector.
        Returns:
            float: Euclidean distance.
        """
        return np.linalg.norm(x - y)

    def train(self, features, labels):
        """
        Train the classifier by initializing or updating prototypes.
        Args:
            features (numpy.ndarray): Feature vectors for training data.
            labels (numpy.ndarray): Corresponding labels for training data.
        """
        for feature, label in zip(features, labels):
            if label not in self.prototypes:
                # Initialize prototype for the class
                self.prototypes[label] = feature
                self.class_counts[label] = 1
            else:
                # Update prototype by incremental averaging
                self.prototypes[label] = (
                    self.prototypes[label] * self.class_counts[label] + feature
                ) / (self.class_counts[label] + 1)
                self.class_counts[label] += 1

    def predict(self, features):
        """
        Predict class labels for given feature vectors.
        Args:
            features (numpy.ndarray): Feature vectors.
        Returns:
            numpy.ndarray: Predicted labels.
        """
        predictions = []
        for feature in features:
            # Find the closest prototype
            distances = {
                label: self.euclidean_distance(feature, prototype)
                for label, prototype in self.prototypes.items()
            }
            # Assign the label of the closest prototype
            predictions.append(min(distances, key=distances.get))
        return np.array(predictions)

    def update_classifier(self, features, pseudo_labels):
        """
        Incrementally update the classifier with new data.
        Args:
            features (numpy.ndarray): Feature vectors.
            pseudo_labels (numpy.ndarray): Pseudo-labels for the data.
        """
        for feature, pseudo_label in zip(features, pseudo_labels):
            if pseudo_label not in self.prototypes:
                # Initialize prototype for the new class
                self.prototypes[pseudo_label] = feature
                self.class_counts[pseudo_label] = 1
            else:
                # Update prototype for the existing class
                self.prototypes[pseudo_label] = (
                    self.prototypes[pseudo_label] * self.class_counts[pseudo_label] + feature
                ) / (self.class_counts[pseudo_label] + 1)
                self.class_counts[pseudo_label] += 1


# ----------------------------------
# Main Code
# ----------------------------------
import pickle

if __name__ == "__main__":
    # Initialize the FeatureExtractor
    feature_extractor = FeatureExtractor(batch_size=32)

    # Initialize the PrototypeClassifier
    classifier = PrototypeClassifier()

    accuracies = []

    for i in range(1, 11):  # Loop through datasets 1 to 10
        # Step 1: Load Training Data
        train_data_path = f'part_one_dataset/train_data/{i}_train_data.tar.pth'
        t = torch.load(train_data_path)  # Load training data
        data = t['data']
        data = data.astype('float32') / 255.0

        # Step 2: Extract Features
        features = feature_extractor.extract(data)

        if i == 1:
            # For the first dataset, use the provided labels and train the classifier
            targets = t['targets']
            classifier.train(features, targets)
        else:
            # For subsequent datasets, predict pseudo-labels using the current classifier
            pseudo_labels = classifier.predict(features)

            # Update the classifier with the new data and pseudo-labels
            classifier.update_classifier(features, pseudo_labels)

        # Save features and labels (real or pseudo) for reference
        np.save(f"D{i}_features.npy", features)
        if i == 1:
            np.save(f"D{i}_labels.npy", targets)
        else:
            np.save(f"D{i}_labels.npy", pseudo_labels)

        # Save the updated classifier as a .pkl file
        with open(f"f{i}.pkl", "wb") as model_file:
            pickle.dump(classifier, model_file)

        # Step 3: Evaluate the Classifier on the Corresponding Evaluation Data
        eval_data_path = f'part_one_dataset/eval_data/{i}_eval_data.tar.pth'
        eval_data = torch.load(eval_data_path)  # Load evaluation data
        eval_images, eval_labels = eval_data['data'], eval_data['targets']
        eval_images = eval_images.astype('float32') / 255.0

        # Extract features for evaluation data
        eval_features = feature_extractor.extract(eval_images)

        # Predict labels for evaluation data
        eval_predictions = classifier.predict(eval_features)

        # Evaluate the predictions
        accuracy = accuracy_score(eval_labels, eval_predictions)
        precision = precision_score(eval_labels, eval_predictions, average="weighted", zero_division=0)
        recall = recall_score(eval_labels, eval_predictions, average="weighted", zero_division=0)
        f1 = f1_score(eval_labels, eval_predictions, average="weighted", zero_division=0)
        accuracies.append((accuracy, precision, recall, f1))

        print(f"Model f{i}: Accuracy={accuracy:.4f}, Precision={precision:.4f}, Recall={recall:.4f}, F1-Score={f1:.4f}")

    # Final Summary
    print("\nFinal Summary of Accuracies:")
    for i, (accuracy, precision, recall, f1) in enumerate(accuracies, start=1):
        print(f"Model f{i}: Accuracy={accuracy:.4f}, Precision={precision:.4f}, Recall={recall:.4f}, F1-Score={f1:.4f}")

```
